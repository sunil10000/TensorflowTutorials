```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist

dir_path = '/content/drive/My Drive/Colab Notebooks/Tensorflow Tutorial'
```

```
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28).astype('float32')/255.0
x_test  =  x_test.reshape(-1, 28*28).astype('float32')/255.0
```

```
model1 = keras.Sequential(
    [
     layers.Dense(64, activation='relu', name='layer1'),
     layers.Dense(10, name='layer2')
    ]
)

model = model1

model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(),
    metrics=['accuracy']
)


model.fit(x_train, y_train, batch_size=64, epochs=5, verbose=2)
model.evaluate(x_test, y_test, batch_size=32, verbose=2)
```

```
print(model.summary())
```

<pre>
                  <b>Saving Model - 2 Methods</b>
1) Save and load model weights 
2) Save and load entire model (Serializing model)
    - Save weights
    - Model architecture
    - Training Configuration (model.compil())
    - Optimizer and states (e.g. exponential weighted averages in adam)
Can export model to production app(javascript etc)
Saved as data structure
</pre>

```
model2 = keras.Sequential(
    [
     layers.Dense(64, activation='relu'),
     layers.Dense(10)
    ]
)
model2.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(),
    metrics=['accuracy']
)
model2.evaluate(x_test, y_test, batch_size=32, verbose=2)
```

<pre>                  <b> Only saving weights </b>
To load weights model architecture shoudl be same and
 also the way model is created
 (functional or sequential or subclassing or custom layer) </pre>

```
model.save_weights(f'{dir_path}/saved_model/') # can use save_format='h5' or tf_checkpoint
model2.load_weights(f'{dir_path}/saved_model/') # automaticaly use correct format
model2.evaluate(x_test, y_test, batch_size=32, verbose=2)
```

<pre>
                      <b> Save whole Model </b>
  This works regradless of architecture,
  sequential, functional, subclassing, custom layer
</pre>

```
model.save(f'{dir_path}/complete_saved_model/')

from keras.models import load_model
model3 = load_model(f'{dir_path}/complete_saved_model/')
model3.fit(x_train, y_train, batch_size=32, epochs=1, verbose=2)
# model3.evaluate(x_test, y_test, batch_size=32, verbose=2)
```
