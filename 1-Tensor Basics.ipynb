```
import tensorflow as tf
```

# Initialization of Tensors

```
x = tf.constant(4, shape=(2, 3), dtype=tf.float32)
print(x)

x = tf.constant([[1,2,3], [4,5,6]])
print(x)

x = tf.ones((3,3))
x = tf.zeros((2,3))
x = tf.eye(3) # I for indentity matrix
print(x)

x = tf.random.normal((3,3), mean=0, stddev=1)
print(x)

x = tf.random.uniform((1,3), minval=0, maxval=1)
print(x)

x = tf.range(9)
x = tf.range(start=2, limit=10, delta=2)
print(x)

x = tf.cast(x, dtype=tf.float32)
print(x)

# tf.float(16,32,64)
# tf.int(8,16,32,64)
# tf.bool
```

# Mathematical Operations

```
x = tf.constant([1, 2, 3])
y = tf.constant([9, 8, 7])

# Equivalent ways
z1 = tf.add(x, y)
z = x + y 
print(z)

z = x - y # tf.subtract
print(z)

z = x*y   # tf.multiply
print(z)

z = x/y   # tf.divide
print(z)

z = x**3
print(z)
```

```
z = tf.tensordot(x, y, axes=1)
print(z)

z = tf.reduce_sum(x*y, axis=0)
print(z)

x = tf.random.normal((2,3))
y = tf.random.normal((3,4))
z = tf.matmul(x, y)
print(z)
z = x @ y
print(z)
```

# Indexing

```
x = tf.constant([0, 1, 2, 3, 4, 5, 6, 7])

print(x[:])
print(x[1:])
print(x[1:3])
print(x[::2])
print(x[::-1])
print(x[5:1:-2]) # start:end:step in [start, end)

indices = tf.constant([0, 3])
x_ind   = tf.gather(x, indices)
print(x_ind)
```

```
x = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])

print(x[0:2, :])
```

# Reshaping

```
x = tf.range(9)
print(x)

x = tf.reshape(x, (3,3))
print(x)

x = tf.transpose(x, perm=[1, 0])
print(x)
```

```
x = tf.reshape(tf.range(27), (3,3,3))
print(x)

print('\n==============================\n')

x = tf.transpose(x, perm=[1, 0, 2])
print(x)

print('\n===============================\n')

x = tf.transpose(x, perm=[2, 1, 0])
print(x)
```
